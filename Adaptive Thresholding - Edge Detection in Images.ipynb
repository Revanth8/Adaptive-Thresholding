```python
import cv2
import matplotlib.pyplot as plt
```

```python
col_img = cv2.imread('House.jpg')
gray_img = cv2.cvtColor(col_img, cv2.COLOR_BGR2GRAY)
```

```python
## 2nd argument - threshold, 3rd Argument - Value assigned, if pixel is greater than threshold
_,th1 = cv2.threshold(gray_img,127,255,cv2.THRESH_BINARY) 

## threshold value is the mean of neighbourhood area
th2 = cv2.adaptiveThreshold(gray_img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,9,2)
th2 = cv2.bitwise_not(th2)

## OTSU thresholding
_,th3 = cv2.threshold(gray_img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
```

```python
## Edge detection using canny algorithm
canny = cv2.Canny(gray_img,100,200)
```

```python
## Create an image which highlights edges
res_img = cv2.bitwise_and(col_img, col_img, mask = canny)
```

```python
## Plot Images
titles = ["Colour", "Gray", "Simple_Thresh", "Adaptive_Thresh", "Otsu_Thresh", "Canny", "Result"]
cmaps = [None, "gray", "gray", "gray", "gray", "gray", None]
images = [col_img, gray_img, th1, th2, th3, canny, res_img]

plt.figure(figsize=(18,12))
for i in range(7):
    plt.subplot(3, 3, i+1)
    plt.imshow(images[i], cmap=cmaps[i])
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])

plt.show()
```
